```python
from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import operator
```

```python
people = fetch_lfw_people(min_faces_per_person = 20, resize=0.7)
image_shape = people.images[0].shape
fig,axes = plt.subplots(2,5,figsize=(15,8),subplot_kw={'xticks':(),'yticks':()})

for target, image, ax in zip(people.target, people.images,axes.ravel()):
    ax.imshow(image, cmap=cm.gray)
    ax.set_title(people.target_names[target])

print("people.image.shape:{}".format(people.images.shape))
print("Number of classes:{}".format(len(people.target_names)))

counts = np.bincount(people.target)
for i, (count,name) in enumerate(zip(counts,people.target_names)):
    print('{0:25} {1:3}'.format(name,count), end =' ')
    if (i+1)%3 == 0:
        print()
```

```python
mask = np.zeros(people.target.shape, dtype=np.bool)
for target in np.unique(people.target) :
    mask[np.where(people.target == target)[0][:50]] = 1
X_people = people.data[mask]
y_people = people.target[mask]
# scale the grayscale values to be between 0 and 1
# instead of 0 and 255 for better numeric stability
X_people = X_people / 255.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_people, y_people, stratify=y_people, random_state=0)

mean = np.mean(X_train, axis=0)
scale = np.std(X_train, axis=0, ddof=1)
X_std = (X_train - mean) / scale
Xte_std = (X_test - mean) / scale
#print(len(X_std))
```

```python
def Eigen(c_mat):
    e_vals, e_vecs = np.linalg.eig(c_mat)
    index = e_vals.argsort()[::-1]
    e_vals = e_vals[index]
    e_vecs = e_vecs[:, index]
    return e_vals, e_vecs


def Eigen_2(c_mat):
    e_vals, e_vecs = np.linalg.eig(c_mat)
    index = e_vals.argsort()[::-1]
    d = np.where(index.cumsum() >= (e_vals.sum() * 0.95))[0][0] + 1
    e_vals2 = e_vals.argsort()[-d:]
    e_vecs2 = e_vecs[:, e_vals2[::-1]]
    return e_vals,e_vecs2

def Eigen_3(c_mat, d):
    e_vals, e_vecs = np.linalg.eig(c_mat)
    e_vals2 = e_vals.argsort()[-d:]
    val = e_vals[e_vals2[::-1]]
    e_vecs2 = e_vecs[:, e_vals2[::-1]]
    return val, e_vecs2

def PCA_2D(e_vecs,X_std):
    e_vec2d = e_vecs[:, :2]
    dot_p2d = np.dot(X_std, e_vec2d)
    return e_vec2d,dot_p2d
```

```python
c_mat = X_std.T @ X_std
e_vals,e_vecs = Eigen(c_mat)
dot_p = np.dot(X_std,e_vecs)
#2D Reduction
e_vec2d,dot_p2d=PCA_2D(e_vecs,X_std)

# def PCA_2D(e_vec2,X_std,d=2):
#     xtr_dim = X_std @ e_vec2
#     #xtr_dim = np.array(xtr_dim).T
#     return xtr_dim

#c_mat = X_std.T @ X_std
# e_vals, e_vecs = np.linalg.eig(c_mat)
# #xtr_dim = PCA_2D(e_vecs,X_std)

#a = np.argmax(e_vals)
#b = np.argmin(e_vals)
#c = xtr_dim[a]
#d= xtr_dim[b]
#print(c)
#print(len(xtr_dim))
#print(b)
#print(d)

#e_val,evecs = Eigen(c_mat)
#x_2D= PCA_2D(evecs,X_std,d=2)
```

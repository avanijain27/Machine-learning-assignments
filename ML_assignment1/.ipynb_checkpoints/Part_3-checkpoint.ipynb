```python
import pandas as pd  
import numpy as np  
from sklearn.model_selection import KFold 
import random
from sklearn.metrics import mean_squared_error
```

```python
#Reading data
matrix = pd.read_csv("/Users/avanijain/Documents/ML_assignment1/x06Simple.csv", usecols = range(1,4))

Data = matrix.iloc[:, 0:4].values
```

```python
Fold_size = 20

np.random.seed(0)
np.random.shuffle(Data)

for i in range (0, 20):
    
    MeanSquareError = []
    
    K_Fold = KFold(n_splits = Fold_size)
    K_Fold.get_n_splits(Data)
    
    for train_index, test_index in K_Fold.split(Data):
               
        train_part, test_part = Data[train_index], Data[test_index]
        
        X1_train = np.array(train_part[:, 0:2])
        X1_test = np.array(test_part[:, 0:2])
        Y1_train = np.array(train_part[:, -1])
        Y1_test = np.array(test_part[:, -1])        
        
        
        X_S_train = (X1_train - np.mean(X1_train)) / np.std(X1_train, ddof = 1)
       
        X_test = (X1_test - np.mean(X1_train)) / np.std(X1_train, ddof = 1)
       
        
        Bias = []
        for i in range(0, len(X_S_train)):
            Bias.insert(i, [1])
        X_S_train = np.insert(X_S_train, [0] , Bias, axis = 1)
        
        Bias = []
        for i in range(0, len(X_test)):
            Bias.insert(i, [1])
        X_test = np.insert(X_test, [0] , Bias, axis = 1)
        
        X_train_transpose = X_S_train.transpose()

        M_Mul1 = np.dot(X_train_transpose, X_S_train)
        M_Mul2 = np.dot(X_train_transpose, Y1_train)
        M_Mul1_inv = np.linalg.inv(M_Mul1)

        Theta = np.dot(M_Mul1_inv,M_Mul2)
        #print(Theta)
        
        #Testing the dataset
        Predict = np.dot(X_test, Theta)
        #print(Predict)
        #print(Y_test)
        
        #Computing the root mean squared error (RMSE)
        MeanSquareError.append(np.sqrt(np.mean((Y1_test - Predict) ** 2))) 
        
    print(MeanSquareError)
    
print("Overall Mean: ", np.mean(MeanSquareError))
print("Overall Standard Deviation: ", np.std(MeanSquareError, ddof = 1))
```

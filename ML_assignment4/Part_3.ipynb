```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
import imageio
```

```python
def convolve(o_img, kernel):
    k_h = kernel.shape[0]
    k_w = kernel.shape[1]
    
    height = math.floor(k_h/2)
    width = math.floor(k_w/2)

    if (len(o_img.shape) == 3):
        img_p = np.pad(o_img, pad_width=((math.floor(k_h/2), math.floor(k_h/2)), \
                                                (math.floor(k_w/2), math.floor(k_w/2)), \
                                                (0, 0)), mode='constant', \
                           constant_values=0).astype(np.float32)
    elif (len(o_img.shape) == 2):
        img_p = np.pad(o_img, pad_width=( \
            (math.floor(k_h/2), math.floor(k_h/2)), (math.floor(k_w/2), math.floor(k_w/2))) \
                           , mode='constant', constant_values=0) \
            .astype(np.float32)

        h_end = -height
        w_end = -width

    img_conv = np.zeros(img_p.shape)
    for i in range(height, img_p.shape[0] - height):
        for j in range(width, img_p.shape[1] - width):
            z = img_p[i - height:i - height + k_h, j - width:j - width + k_w]
            z = z.flatten() * kernel.flatten()
            img_conv[i][j] = z.sum()


    if (height == 0):
        return img_conv[height:, width:w_end]
    if (width == 0):
        return img_conv[height:h_end, width:]
    return img_conv[height:h_end, width:w_end]

# def gaussian_formula(x, y, sd):
#     return 1 / (2*np.pi*(sd**2)) * np.exp(-(x**2 + y**2)/(2* sd**2))
```

```python
def Gaussiannorm(sigma):
    f_size = int(4 * sigma + 0.5) + 1
    gauss_f = np.zeros((f_size, f_size), np.float32)
    u = math.floor(f_size/2)
    v = math.floor(f_size/2)
    for x in range(-u, u + 1):
        for y in range(-v, v + 1):
            formula = 1 / (2*np.pi*(sigma**2)) * np.exp(-(x**2 + y**2)/(2* sigma**2))
            gauss_f[x + u, y + v] = formula
    return gauss_f
```

```python
def GaussianBlurImage(image, sigma):
    gauss_f = Gaussiannorm(sigma)
    i_filter = np.zeros_like(image, dtype=np.float32)
    for c in range(3):
        i_filter[:, :, c] = convolve(image[:, :, c], gauss_f)
    return (i_filter.astype(np.uint8))
```

```python
image = imageio.imread("./color1.jpg")
plt.figure( figsize = (10,10))
plt.imshow(image)
plt.title("Original Image")
plt.show()

image = GaussianBlurImage(image, 1)
plt.figure( figsize = (10,10))
plt.imshow(image,vmin=0, vmax=255)
plt.title("Gaussian Blurred Image")
plt.show()
```

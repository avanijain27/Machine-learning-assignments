```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
```

```python
def convolution(image, kernel):

    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image_row, image_col = image.shape
    kernel_h, kernel_w = kernel.shape
    output = np.zeros(image.shape)

    padded_image = np.pad(image, pad_width=((math.floor(kernel_h / 2), math.floor(kernel_h / 2)), (math.floor(kernel_w / 2), math.floor(kernel_w / 2))), mode='constant', constant_values=0).astype(np.float32)
    for row in range(image_row):
        for col in range(image_col):
            x = padded_image[row:row + kernel_h, col:col + kernel_w]
            x = x.flatten() * kernel.flatten()
            output[row][col] = x.sum()

    return output

```

```python
def showImg(img, title, vmin=0, vmax=255):
    plt.imshow(img, cmap='gray',vmin=vmin,vmax=vmax)
    plt.title(title)
    plt.show()
```

```python
def sobel_edge_detection(image, filter):
    gx = convolution(image, filter)
    gy = convolution(image, np.flip(filter.T, axis=0))
    gradient_magnitude = np.sqrt(np.square(gx) + np.square(gy))
    gradient_magnitude *= 255.0 / gradient_magnitude.max()
    showImg(gx, "Horizontal Edge")
    showImg(gy, "Vertical Edge")
    # showImg(gradient_magnitude, "gradient_magnitude")
    img_threshold = threshold(gradient_magnitude, 10, 25, 50)
    return img_threshold
```

```python
def threshold(image,lower, upper,t):

    output = np.zeros(image.shape)
    s_row, s_col = np.where(image >= upper)
    t_row, t_col = np.where((image <= upper) & (image >= lower))
    output[s_row, s_col] = 255
    output[t_row, t_col] = t
    return output
```

```python
filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
image = cv2.imread("./color1.jpg")
img=sobel_edge_detection(image, filter)
showImg(img, "Gradient Magnitude", 0, 255)
```

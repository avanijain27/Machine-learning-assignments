```python
#GRAYSCALE

import numpy as np
import imageio
import matplotlib.pyplot as plt
%matplotlib inline

pic = imageio.imread("./color1.jpg")
#print(pic)

r = pic[ : , : , 0]
g = pic[ : , : , 1]
b = pic[ : , : , 2]
gray1 = np.dot(r,0.2989) + np.dot(g,0.5870) + np.dot(b,0.1140)
plt.figure( figsize = (10,10))
plt.imshow(gray1, cmap = plt.get_cmap(name = 'gray'))
plt.show()
```
